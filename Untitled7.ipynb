```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def downsmple_im(inpt_im_path, smpling_factor):

    inpt_im = cv2.imread(inpt_im_path)

    new_width = inpt_im.shape[1] // smpling_factor
    new_height = inpt_im.shape[0] // smpling_factor
    downsmpled_im = cv2.resize(inpt_im, (new_width, new_height), interpolation=cv2.INTER_NEAREST)

    return inpt_im, downsmpled_im

def display_images(original_im, downsmpled_im):

    original_im = cv2.cvtColor(original_im, cv2.COLOR_BGR2RGB)
    downsmpled_im = cv2.cvtColor(downsmpled_im, cv2.COLOR_BGR2RGB)
    # original image
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(original_im)
    plt.axis('off')
    # downsampled image
    plt.subplot(1, 2, 2)
    plt.title('Downsampled Image')
    plt.imshow(downsmpled_im)
    plt.axis('off')

    plt.show()
```

```python
inpt_im_path = "/content/H.webp"
smpling_factor = 2

original_im, downsmpled_im = downsmple_im(inpt_im_path, smpling_factor)

```

```python
# Display the images
display_images(original_im, downsmpled_im)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def qntize_im(inpt_im_path, qntization_levels):

    inpt_im = cv2.imread(inpt_im_path)

    # Converting image to float32 for accurate calculations
    im_float = inpt_im.astype(np.float32)

    qntization_step = 256 // qntization_levels
    qntized_im = np.floor(im_float / qntization_step) * qntization_step

    # Converting back to uint8 for visualization
    qntized_im = qntized_im.astype(np.uint8)

    return inpt_im, qntized_im

def displ_im(original_im, qntized_im):
    # Converting BGR to RGB for Matpltlib
    original_im = cv2.cvtColor(original_im, cv2.COLOR_BGR2RGB)
    qntized_im = cv2.cvtColor(qntized_im, cv2.COLOR_BGR2RGB)

    # original image
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(original_im)
    plt.axis('off')

    #quantized image
    plt.subplot(1, 2, 2)
    plt.title('Quantized Image')
    plt.imshow(qntized_im)
    plt.axis('off')

    plt.show()
```

```python
inpt_im_path = '/content/H.webp'
qntization_levels = 16
original_im, quantized_im = qntize_im(inpt_im_path, qntization_levels)
```

```python
displ_im(original_im, quantized_im)
```
