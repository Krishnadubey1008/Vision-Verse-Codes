gaussian filtering on first image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_gaussian_filter(img_path, kernel_size):

    img = cv2.imread(img_path)

    # Converting image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Applying Gaussian blur
    blurred_img = cv2.GaussianBlur(gray_img, (kernel_size, kernel_size), 0)

    # Displaying the original and filtered images
    plt.subplot(121), plt.imshow(gray_img, cmap='gray'), plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(blurred_img, cmap='gray'), plt.title('Gaussian Filtered')
    plt.xticks([]), plt.yticks([])

    plt.show()


    cv2.imwrite('gaussian_filtered_img.jpg', blurred_img)

# Specify the path to the image and the kernel size for the Gaussian filter
img_path = '/content/lowContrast.png'
kernel_size = 5  # Adjust the kernel size as needed

# Apply the Gaussian filter
apply_gaussian_filter(img_path, kernel_size)
```

Laplacian filter on first image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_laplacian_filter(img_path):
    # Read the image
    img = cv2.imread(img_path)

    # Converting image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Applying Laplacian filter
    laplacian_filtered_img = cv2.Laplacian(gray_img, cv2.CV_64F)

    # Taking absolute value of result
    laplacian_filtered_img = np.abs(laplacian_filtered_img)

    # Displaying the original and filtered images
    plt.subplot(121), plt.imshow(gray_img, cmap='gray'), plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(laplacian_filtered_img, cmap='gray'), plt.title('Laplacian Filtered')
    plt.xticks([]), plt.yticks([])

    plt.show()

    # Saveing filtered image
    cv2.imwrite('laplacian_filtered_image.jpg', laplacian_filtered_img)


img_path = '/content/lowContrast.png'

# Applying Laplacian filter
apply_laplacian_filter(img_path)
```

Bilateral filter on first image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def apply_bilateral_filter(img_path, d, sigma_color, sigma_space):

    img = cv2.imread(img_path)

    # Applying bilateral filter
    bilateral_filtered_img = cv2.bilateralFilter(img, d, sigma_color, sigma_space)

    # Displaying original and filtered images
    plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(cv2.cvtColor(bilateral_filtered_img, cv2.COLOR_BGR2RGB)), plt.title('Bilateral Filtered')
    plt.xticks([]), plt.yticks([])

    plt.show()


    cv2.imwrite('bilateral_filtered_image.jpg', bilateral_filtered_img)

img_path = '/content/lowContrast.png'


d = 9  # Diameter of each pixel neighborhood
sigma_color = 75  # Filter sigma in the color space
sigma_space = 75  # Filter sigma in the coordinate space

# Applying bilateral filter
apply_bilateral_filter(img_path, d, sigma_color, sigma_space)
```
